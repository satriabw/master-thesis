```python
import matplotlib.pyplot as plt
import pandas as pd
from pymeos import *

pymeos_initialize()
```

```python
ais = pd.read_csv('./data/aisdk-2023-10-19.csv')

ais['# Timestamp'] = pd.to_datetime(ais['# Timestamp'], format='%d/%m/%Y %H:%M:%S')
ais['# Timestamp'] = ais['# Timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S')

ais.head()
```

```python
filtered_ais = ais.copy()
# filtered_ais = ais[ais["MMSI"] == 2182807]
```

```python
filtered_ais['point'] = filtered_ais.apply(lambda row: TGeogPointInst(point=(row['Latitude'], row['Longitude']), timestamp=row['# Timestamp']),
                         axis=1)
filtered_ais['sog'] = ais.apply(lambda row: TFloatInst(value=row['SOG'], timestamp=row['# Timestamp']), axis=1)
filtered_ais.drop(['Latitude', 'Longitude', 'Type of mobile', 'Callsign', 'IMO', 'Name', 'Ship type', 'Cargo type', 'Width', 'Navigational status', 'ROT', 'COG', 'Heading', 'Length', 'Type of position fixing device', 'Draught', 'Destination', 'Data source type', 'ETA', 'A', 'B', 'C', 'D'], axis=1, inplace=True)

filtered_ais.head()
```

```python
trajectories = filtered_ais.groupby('MMSI').aggregate(
    {
        'point': TGeogPointSeq.from_instants,
        'sog': TFloatSeq.from_instants
    }
).rename({'point': 'trajectory'}, axis=1)
trajectories['distance'] = trajectories['trajectory'].apply(lambda t: t.length())
trajectories.head()
```

```python
pd.concat([filtered_ais.groupby('MMSI')['# Timestamp'].count().rename('original #points'),
           trajectories['trajectory'].apply(lambda t: t.num_instants()).rename('PyMEOS #points')],
          axis=1)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))
for _, ship in trajectories.iterrows():
    ship['trajectory'].plot(axes=axes[0])
    ship['sog'].plot(axes=axes[1])
plt.show()
```

```python
pymeos_finalize()
```
